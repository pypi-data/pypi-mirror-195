```python
## ACME autophase routines, keep collapsed
def autops(data, fn, p0=0.0, p1=0.0):
    """
    Automated phase correction from NMRglue by https://github.com/jjhelmus
    These functions provide support for automatic phasing of NMR data. 


    Automatic linear phase correction

    Parameters

        data : ndarray

             Array of NMR data.

        fn : str or function

             Algorithm to use for phase scoring. Built in functions can be
             specified by one of the following strings: "acme", "peak_minima"

        p0 : float

            Initial zero order phase in degrees.

        p1 : float

            Initial first order phase in degrees.

    Returns

        ndata : ndarray

            Phased NMR data.

    """

    import numpy as np
    import scipy.optimize
    from io import StringIO # Python3 use: from io import StringIO
    from contextlib import redirect_stdout

    
    if not callable(fn):
        fn = {
            'peak_minima': _ps_peak_minima_score,
            'acme': _ps_acme_score,
        }[fn]
    
    opt = [p0, p1]
    with StringIO() as buf, redirect_stdout(buf):   
        opt = scipy.optimize.fmin(fn, x0=opt, args=(data, ))
        mystdout = buf.getvalue()
    return ps(data, p0=opt[0], p1=opt[1]), opt[0], opt[1], mystdout


def _ps_acme_score(ph, data):
    """
    Phase correction using ACME algorithm by Chen Li et al.
    Journal of Magnetic Resonance 158 (2002) 164-168

    Parameters
    * pd : tuple, current p0 and p1 values
    * data : ndarray, array of NMR data.

    Returns
    * score : float, value of the objective function (phase score)

    """
    import numpy as np

    stepsize = 1

    phc0, phc1 = ph

    s0 = ps(data, p0=phc0, p1=phc1)
    data = np.real(s0)

    # Calculation of first derivatives
    ds1 = np.abs((data[1:]-data[:-1]) / (stepsize*2))
    p1 = ds1 / np.sum(ds1)

    # Calculation of entropy
    p1[p1 == 0] = 1

    h1 = -p1 * np.log(p1)
    h1s = np.sum(h1)

    # Calculation of penalty
    pfun = 0.0
    as_ = data - np.abs(data)
    sumas = np.sum(as_)

    if sumas < 0:
        pfun = pfun + np.sum((as_/2) ** 2)

    p = 1000 * pfun

    return h1s + p


def _ps_peak_minima_score(ph, data):
    """
    Phase correction using simple minima-minimisation around highest peak
    This is a naive approach but is quick and often achieves reasonable
    results.  The optimisation is performed by finding the highest peak in the
    spectra (e.g. TMSP) and then attempting to reduce minima surrounding it.
    Parameters
    * pd : tuple, current p0 and p1 values
    * data : ndarray, array of NMR data.

    Returns
    * score : float, value of the objective function (phase score)

    """

    phc0, phc1 = ph

    s0 = ps(data, p0=phc0, p1=phc1)
    data = np.real(s0)

    i = np.argmax(data)
    mina = np.min(data[i-100:i])
    minb = np.min(data[i:i+100])

    return np.abs(mina - minb)

def ps(data, p0=0.0, p1=0.0, inv=False):
    """
    Linear phase correction

    Parameters

        data : ndarray

            Array of NMR data.

        p0 : float

            Zero order phase in degrees.

        p1 : float

            First order phase in degrees.

        inv : bool, optional

            True for inverse phase correction

    Returns

        ndata : ndarray

            Phased NMR data.

    """
    import numpy as np

    p0 = p0 * np.pi / 180.  # convert to radians
    p1 = p1 * np.pi / 180.
    size = data.shape[-1]
    apod = np.exp(1.0j * (p0 + (p1 * np.arange(size) / size))).astype(data.dtype)
    if inv:
        apod = 1 / apod
    return apod * data
```

```python
# %matplotlib -l  lists all the possible options
import numpy as np
import matplotlib.pyplot as P

# generate 8 detector count rates, GPS like
N0,dt,taumu,gamu = 1000, 0.5e-3, 2.197.,2*np.pi*135.5 # ref detector rate in counts/bin, bin width in mus, muon lifetime in mus, gyromagnetic ratio in mus-1/T 
ag,wg,sg = np.array([0.15,0.2,0.3]),gamu*np.array([.35,0.31,0.37]),np.array([0.8,3.2,1.6]) # Gaussian amplitudes sum to 0.6,omega as gamu*field in Tesla, decays in us-1
al,wl,ll = np.array([0.18,0.12,0.05]), gamu*np.array([0.3,0.36,0.39]),np.array([0.3,1.5,2.5]) # Lorentzian amplitudes sum to 0.4
t0 =  np.zeros(8) # 0.0015 + np.array([0.0006,-0.0008,-0.0004,-0.0006,-0.0007,0.0008,0.0004,0.0007])  # errors in t0 placement of each detector, average is 0.22 # np.zeros(8) # 
alpha = np.array([1.,0.95,0.93,0.97,1.03,1.05,1.04,0.99])# ratios of N0/Ni
theta = np.linspace(0,7/8,8)*2*np.pi # exact angles! check np.cos(theta).sum()=-4-4e-16
dtheta = np.array([0,-1,-1,1,-1,0,1,1,])*np.pi/64
theta = theta + dtheta # sum = 0
# detectors along columns:
t0 = np.reshape(t0,(8,1))
alpha = np.reshape(alpha,(8,1))
theta = np.reshape(theta,(8,1))

print(r'$\sum_i \cos\theta_i =$ {:.2e}'.format(np.cos(theta).sum()))
t = np.arange(0.,10.,dt)  # nominal-time vector along rows in musec

tt, tt0 = np.meshgrid(t,t0)
tt, aalpha = np.meshgrid(t,alpha)
tt, ttheta = np.meshgrid(t,theta)

N = sum([N0/aalpha*np.exp(-(tt-tt0)/taumu)*(1.+ag[k]*np.cos(wg[k]*(tt-tt0)+ttheta)*np.exp(-0.5*(sg[k]*(tt-tt0))**2)+al[k]*np.cos(wl[k]*(tt-tt0)+ttheta)*np.exp(-ll[k]*(tt-tt0))) for k in range(ag.shape[0])],0) # unpolarized counts for all detectors
# 8 detectors counts
Px = sum([ag[k]*np.cos(wg[k]*(t-t0[0]))*np.exp(-0.5*(sg[k]*(t-t0[0]))**2)+al[k]*np.cos(wl[k]*(t-t0[0]))*np.exp(-ll[k]*(t-t0[0])) for k in range(ag.shape[0])],0)
# pure cos polarization




last = 20000
fix,ax = P.subplots(2,4,figsize=(14,7),sharex=True,sharey=True)
for det in range(theta.shape[0]):
    ax[divmod(det,4)].plot(t[:last],N[det,:last],label='# {}'.format(det))
    ax[divmod(det,4)].legend()
    if det>3:
        ax[divmod(det,4)].set_xlabel(r't [$\mu$s]')
    if det % 4 == 0:
        ax[divmod(det,4)].set_ylabel('Polarization')
P.show()
```

```python
# fake analysis 
Ntot=8*N0*.np.exp(-tt/taumu)
Pc = 2*sum(aalpha*N*np.cos(ttheta),0)/Ntot
Ps = 2*sum(aalpha*N*np.sin(ttheta),0)/Ntot
fig,ax = P.subplots(1,3,figsize=(10,5),sharey=True)


fftP = np.fft.fft(Pc)+1.0j*np.fft.fft(Ps)
fftPrephase,p0,p1,dum = autops(fftP,'acme')
print(p0,p1) #'p0 = {:.1f} deg, p1 = {:.1f} deg',format(p0,p1)
Prephase = np.fft.ifft(fftPrephase)
ax[0].plot(t[:last],Pc[:last],label='cos')
ax[0].plot(t[:last],Ps[:last],label='sin')
ax[1].plot(t[:last],Px[:last],label='original')
ax[2].plot(t[:last],Prephase.real[0:last],label='acme cos')
ax[2].plot(t[:last],Prephase.imag[0:last],label='acme sin')
ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[0].set_xlabel(r't [$\mu$s]')
ax[0].set_ylabel('Polarization')
ax[1].set_xlabel(r't [$\mu$s]')
ax[2].set_xlabel(r't [$\mu$s]')
ax[0].set_ylim(-1.05,1.05)
P.show()
```
